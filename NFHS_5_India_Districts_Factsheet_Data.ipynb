```python
state_sex_ratio = df.groupby('State')['Sex_Ratio'].mean().reset_index()
state_sex_ratio.columns = ['State', 'Avg_Sex_Ratio']

# 1000 se zyada = females zyada, kam = males zyada
state_sex_ratio['Dominance'] = state_sex_ratio['Avg_Sex_Ratio'].apply(
    lambda x: '👩 Females More' if x > 1000 else '👨 Males More'
)

state_sex_ratio.sort_values('Avg_Sex_Ratio', ascending=False).head(5)
```
