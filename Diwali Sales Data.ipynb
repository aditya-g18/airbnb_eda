```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#to show the graph in the notebook
%matplotlib inline    
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ADITYA GUPTA\Downloads\Data\Diwali Sales Data.csv", encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df['Gender'].dtype
```

```python
df['Gender'] = df['Gender'].str.strip().str.upper()
```

```python
df[['Gender']].head(10)
```

```python
df['Gender'] = df['Gender'].map({'M': 'Male', 'F': 'Female'}) #to replace the values in the Gender
df[['Gender']].head(10)
```

```python
df[['Marital_Status']].head(10)
```

```python
df['Marital_Status'] = df['Marital_Status'].astype(bool) #to convert the Marital_Status column to boolean type, where True represents married and False represents unmarried
df[['Marital_Status']].head(10)
```

```python
df.head(5)
```

```python
df.drop(columns = {'Status', 'unnamed1'}, inplace = True) #to drop the Status and unnamed1 columns from the dataframe
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.value_counts()
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].fillna(df['Amount'].mean()) #to fill the missing values in the Amount column with the mean of the column
df.isnull().sum()
```

```python
df.info()
```

```python
df.head(5)
```

```python
plt.Figure(figsize=(6,3))

plt.bar(df['Zone'], df['Amount'], color = 'lightblue')
plt.xlabel('Zone')
plt.ylabel('Amount')
plt.title('Sales by Zone')

plt.show()
```
