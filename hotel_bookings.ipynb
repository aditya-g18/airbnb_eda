# Data Analysis And Visualaization

```python
df['is_canceled'].unique()
```

```python
df.head(5)
```

```python
df['is_canceled'].value_counts()
```

```python
Cancelled_perc = df['is_canceled'].value_counts(normalize = True) * 100
print(Cancelled_perc)

plt.figure(figsize = (6,4))
plt.title('Distribution of Cancellations')
plt.bar(['Not Cancelled', 'Cancelled'], Cancelled_perc.values, color=['lightblue', 'orange'], edgecolor=['blue', 'orange'], width=0.6)
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
plt.title('Cancellation Distribution by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.legend(title='Cancellation', labels=['Not Canceled', 'Canceled'], bbox_to_anchor=(1.05, 1), loc='upper left')
```
