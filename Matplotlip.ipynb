# What is matplotlib ?
# Matplotlib is a python libeary used for creating static, animated, and intreactive plots.

# Data visualization: its the graphical representation of data to identify patterns, trends and insights

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x,y) # Line plot
plt.show()
```

# Customize chart

```python
plt.figure(figsize=(8, 5)) # Set figure size

plt.title("Line Chart") # Set title
plt.xlabel("X-axis") # Set x-axis label
plt.ylabel("Y-axis") # Set y-axis label
plt.plot(x, y, marker='o', linestyle=':', color='r', linewidth=2, markersize=12) # Line plot with markers and custom style
```

# Advance line chart Multiple line and legend

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 8, 27, 64, 125]

plt.plot(x, y1, label = 'Sales 2024')
plt.plot(x, y2, label = 'Sales 2026')

plt.title("Line Chart with Legend")
plt.xlabel('Months')
plt.ylabel('Sales')

plt.legend() # Show legend
plt.show()
```

# Bar chart

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [2, 4, 9, 16, 25]

plt.bar(x, y) # Bar chart
plt.title("Bar Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.bar(x, y, color='lightblue', edgecolor='blue') # Bar chart with custom colors

plt.show()
```

# Histogram which is used for distribution

```python
import random
data = [random.randint(1, 10) for _ in range(50)] # Generate random data

plt.hist(data, bins=10, color='lightgreen', edgecolor='black') # Histogram
plt.title("Histogram")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Pie chart

```python
Categories = ['A', 'B', 'C', 'D', 'E']
Sales = [20, 35, 30, 35, 27]

plt.pie(Sales, labels=Categories, autopct='%1.1f%%', startangle=90) # Pie chart
plt.title("Pie Chart")
plt.axis('equal') # Equal aspect ratio ensures that pie chart is circular
plt.show()

plt.savefig('pi_chart.png') # Save the figure as a file
```

# Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 9, 16, 25]

plt.scatter(x, y) # Scatter plot
plt.title("Scatter Plot")
```

# Subplots

```python
#data1
Categories = ['Music', 'Books', 'Clothing', 'Electronics', 'Home']
Sales = [20, 35, 30, 35, 27]

#data2
y1 = [20, 35, 30, 35, 27]
y2 = [25, 30, 35, 40, 45]

plt.figure(figsize = (8, 4))

# Bar chart for data1
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st subplot
plt.bar(Categories, Sales, color='lightblue', edgecolor='blue')
plt.title("Sales by Category")
plt.xlabel('Category')
plt.ylabel('Sales')

# Scatter plot for data2
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd subplot
plt.scatter(Categories, y2, color='red')
plt.title("Data2 Scatter Plot")
plt.xlabel('Category')
plt.ylabel('Values')
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
```

# Matplotlip with pandas - real data

```python
# data
data = {
    'Months': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [20, 35, 30, 35, 27]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.bar(df['Months'], df['Sales'], color='lightblue', edgecolor='blue') # Bar chart
plt.title("Sales by Month")
plt.xlabel('Months')
plt.ylabel('Sales')

plt.show()
```
