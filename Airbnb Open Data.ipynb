```python
unique_names = df['host id'].nunique()
print(f'number of unique host id: {unique_names}')
print(f' % of unique host id: {round(unique_names / df["host id"].shape[0] * 100, 2)}%')
```

```python
# Most host IDs are unique, so the 'host id' column is not useful for analysis and can be dropped.
# and dont not need 'id' and 'license' columns as well, so we can drop them too.
df = df.drop(columns=['host id', 'id', 'license'])
```

```python
df.shape
```

```python
unique_names = df['host name'].nunique()
print(f'number of unique host names: {unique_names}')
print(f'% of unique names: {round(unique_names / df["host name"].shape[0] * 100, 2)}%')
print(f'{round(unique_names / df["host name"].shape[0] * 100, 2)}% - that means we have new hosts in this year')
```

```python
percentage_identity_verified = df['host_identity_verified'].value_counts()
print(f'{round(percentage_identity_verified / df["host_identity_verified"].shape[0] * 100, 2)}%')
print(f'\n that means 50% of unconfirmeded hosts:')
```

```python
df['neighbourhood group'].value_counts()
```

```python
#case fixing
df['neighbourhood group'] = df['neighbourhood group'].str.title()
#spelling fixing
df['neighbourhood group'] = df['neighbourhood group'].replace({'Manhattan': 'Manhattan', 'Brooklyn': 'Brooklyn', 'Queens': 'Queens', 'Bronx': 'Bronx', 'Staten Island': 'Staten Island'})
```

```python
df['neighbourhood group'].unique()
```

```python
# now we have to drop nan values from 'neighbourhood group' column
df['neighbourhood group'].isna().sum()
```

```python
#we found 29 nans in 'neighbourhood group' column, we can fill them with 'Unknown' value
df['neighbourhood group'] = df['neighbourhood group'].fillna('Unknown')
```

```python
df['neighbourhood group'].unique()
```

```python
#hmm, lets see how many unique values we have in 'neighbourhood' column
unique_neighbourhoods = df['neighbourhood'].unique().shape[0]
print(f'number of unique neighbourhoods: {unique_neighbourhoods}')
print(f'lets see % of unique neighbourhoods: {round(unique_neighbourhoods / df["neighbourhood"].shape[0] * 100, 2)}%')
```

```python
# we have to change $ sign becoz its not a numeric value and it will creat problem in analysid
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
# converting str to datetime
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')
df['last review'].sort_values(ascending=True).head(7)
```
